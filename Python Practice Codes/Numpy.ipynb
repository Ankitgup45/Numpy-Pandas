```python
import numpy as np
import pandas as pd

data = {
    'Name': ['Ankit', 'Rahul', 'Sonia', 'Raju','Bheem'],
    'Marks/100': [85, 34, 95, 36, 80]
}

df = pd.DataFrame(data)

df['Result'] = np.where(df['Marks/100'] < 35, 'Re-Exam',
                        np.where(df['Marks/100'] > 90, 'Topper', 'Pass'))
print(df)
```
