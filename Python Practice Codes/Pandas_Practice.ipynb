```python
#dataframe using dictionary of lists
import pandas as pd 

data = {
    'Stu_ID': [1, 2, 3, 4, 5],
    'Name': ['Ankit', 'Rahul', 'Sonia', 'Priya', 'Vikram'],
    'Marks/100': [30, 36, 24, 92, 88],
    'Age': [23, 24, 22, 25, 26],
    'City': ['Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Hyderabad'],
    'Pass or Fail': ['Pass', 'Pass', 'Pass', 'Pass', 'Pass']
}
df = pd.DataFrame(data)
print(df)
```

```python
#simple if-else condition
import pandas as pd

data = {
    'Stu_ID': [1, 2, 3, 4, 5],
    'Name': ['Ankit', 'Rahul', 'Sonia', 'Priya', 'Vikram'],
    'Marks/100': [30, 36, 24, 92, 88],
    'Age': [23, 24, 22, 25, 26],
    'City': ['Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Hyderabad']
}

df = pd.DataFrame(data)

# Define condition function
def result_category(marks):
    if marks < 35:
        return "Fail"
    elif marks > 90:
        return "Topper"
    else:
        return "Pass"

# Apply function
df['Result'] = df['Marks/100'].apply(result_category)

print(df)
```

```python
#conditional assignment using np.where
import pandas as pd
import numpy as np

data = {
    'Stu_ID': [1, 2, 3, 4, 5],
    'Name': ['Ankit', 'Rahul', 'Sonia', 'Priya', 'Vikram'],
    'Marks/100': [30, 36, 24, 92, 88],
    'Age': [23, 24, 22, 25, 26],
    'City': ['Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Hyderabad']
}

df = pd.DataFrame(data)

df['Result'] = np.where(df['Marks/100'] < 35, 'Re-Exam',
                np.where(df['Marks/100'] > 90, 'Topper', 'Pass'))

print(df)
```
