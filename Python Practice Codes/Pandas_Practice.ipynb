```python
#simple if-else condition
import pandas as pd

data = {
    'Stu_ID': [1, 2, 3, 4, 5],
    'Name': ['Ankit', 'Rahul', 'Sonia', 'Priya', 'Vikram'],
    'Marks/100': [30, 36, 24, 92, 88],
    'Age': [23, 24, 22, 25, 26],
    'City': ['Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Hyderabad']
}

df = pd.DataFrame(data)

# Define condition function
def result_category(marks):
    if marks < 35:
        return "Fail"
    elif marks > 90:
        return "Topper"
    else:
        return "Pass"

# Apply function
df['Result'] = df['Marks/100'].apply(result_category)

print(df)
```

```python
#conditional assignment using np.where
import pandas as pd
import numpy as np

data = {
    'Stu_ID': [1, 2, 3, 4, 5],
    'Name': ['Ankit', 'Rahul', 'Sonia', 'Priya', 'Vikram'],
    'Marks/100': [30, 36, 24, 92, 88],
    'Age': [23, 24, 22, 25, 26],
    'City': ['Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Hyderabad']
}

df = pd.DataFrame(data)

df['Result'] = np.where(df['Marks/100'] < 35, 'Re-Exam',
                np.where(df['Marks/100'] > 90, 'Topper', 'Pass'))

print(df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("students_practice.csv")
df['result'] = np.where(df['Marks'] < 35, 'Re-Exam',
               np.where(df['Marks'] > 90, 'Topper', 'Pass'))

# Bar chart: each student's marks
df.plot(x='Name', y='Marks', kind='bar')
plt.title("Marks per student")
plt.show()

# Histogram: how marks are spread out
df['Marks'].plot(kind='hist', bins=10)
plt.show()

# Count how many landed in each result category
df['result'].value_counts().plot(kind='bar')
plt.show()
```

```python
#to sort the dataframe based on the values in the 'Bananas' column in ascending order
import pandas as pd

data = { 'Customer_Name': ['Ankit', 'Bheem', 'Priya', 'Raju'],
         'Oranges': [3, 2, 0, 1],
         'Apples': [0, 3, 7, 2],
         'Bananas': [2, 1, 3, 0] }
df = pd.DataFrame(data) 
Sorted = df.sort_values(['Bananas'])# to sort the dataframe based on the values in the 'Bananas' column in ascending order
print(Sorted)
```

```python
#to sort the dataframe based on the values in the 'Oranges' column in descending order
import pandas as pd 
data = { 'Customer_Name': ['Ankit', 'Bheem', 'Priya', 'Raju'],
         'Oranges': [3, 2, 0, 1],
         'Apples': [0, 3, 7, 2],
         'Bananas': [2, 1, 3, 0] }
df = pd.DataFrame(data)# to create a dataframe from the given dictionary of lists     
sorted_df = df.sort_values(by=['Oranges'], ascending=False) # to sort the dataframe based on the values in the 'Oranges' column in descending order
print(sorted_df)
```
