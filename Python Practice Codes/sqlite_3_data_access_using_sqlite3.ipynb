```python
import sqlite3

# Create a connection to the SQLite database
conn = sqlite3.connect('example.db')    
# Create a cursor object to interact with the database
cursor = conn.cursor()
# Create a table named 'users' if it doesn't exist
cursor.execute('''CREATE TABLE IF NOT EXISTS users
                  (id INTEGER PRIMARY KEY, name TEXT UNIQUE, age INTEGER UNIQUE)''')
# Insert a new user into the 'users' table  
cursor.execute("INSERT INTO users (name, age) VALUES (?, ?)", ('Alice', 30))
# Commit the changes to the database
conn.commit()
# Query the 'users' table to retrieve all users
cursor.execute("SELECT * FROM users")
# Fetch all results from the executed query
users = cursor.fetchall()
# Print the retrieved users
print(users)
# Close the database connection
conn.close()  

```

```python
name = input("Enter name: ")

special_chars = "!@#$%^&*()_+=-`~"

if any(char in special_chars for char in name):
    print("Name contains special characters.", name)

elif len(name) < 3:
    print("Name is too short.", name)

elif len(name) > 20:
    print("Name is too long.", name)

else:
    print("Name is valid.", name)
```

```python
#Creating DB
import sqlite3 
conn = sqlite3.connect("Employees.db")
cursor = conn.cursor()

#Creating Table
cursor.execute("""
                       Create table Employees1(
                                                        Emp_Id INTEGER PRIMARY KEY AUTOINCREMENT,
                                                         Emp_Name TEXT,
                                                         Emp_Age INTEGER,
                                                         Emp_Department TEXT,
                                                         Emp_Salary INTEGER
                                                          )
                          """)
print("Table created successfully ")

#Inserting multiple records

data = [
            (1,"Ankit", 23,"AIA AISQUAD",100),
            (2,"Ram", 22,"Life Science",20),
            (3,"Syham", 34,"AIA ",10),
            (4,"Raju", 89,"Insurance",9000),
            (5,"Bheem", 35," AISQUAD",10000),
            (6,"Karan", 45,"Cyber Security",28800),
            (7,"Salman", 70,"Banking",1300),
             (8,"Sharukhan", 45,"Finance",800)]

cursor.executemany("INSERT INTO Employees VALUES (?,?,?,?,?)" , data)

print("Records inserted successfully")
conn.commit()
conn.close()

```

```python
# searching a word
import re

pattern = "My name is raju sarkar , raju bhai have 2 apples and 3 oranges"
x = re.search("raju", pattern)
if x:
    print("Word found in the string.")  
else:
    print("Word not found in the string.")    


```

```python
#finding all the words in the string
import re
pattern = "My name is raju sarkar , raju bhai have 2 apples and 3 oranges"
x = re.findall("raju", pattern)
print(x)
```

```python
#finding digits in a string
import re
pattern = "My name is raju sarkar , raju bhai have 2 apples and 3 oranges"
x = re.findall(r"\d", pattern)
print(x)
```

```python
#Substituting a digits in a string
import re
pattern = "My name is raju sarkar , raju bhai have 2 apples and 3 oranges"
new_pattern = re.sub(r"\d+", "X", pattern)
print(new_pattern)
```

```python
#Substituting a word in a string
import re 
pattern = "My name is raju sarkar , raju bhai have 2 apples and 3 oranges"
new_pattern = re.sub("raju", "Ankit", pattern)
print(new_pattern)  
```

```python
#split function in regular expression
import re 
pattern = "My name is raju sarkar , raju bhai have 2 apples and 3 oranges"
x = re.split(r"\s", pattern)
print(x)
```

```python
#reverse a string
word = "Ankit"
reversed_word = word[::-1]
print(reversed_word)
```

```python
#list comprehension using if condition 
easy = [ X ** 3 for X in range(10) if X % 2 == 0]
print("Cube roots of even numbers from 0 to 9:",easy)
```

```python
#list comprehension using if else condition
Check = [f"{x} =  Even" if x % 2 == 0 else f"{x} = Odd" for x in range(10)]
print(Check)
```
